## Wat is progameren

Een computer programma is een set van instructies die worden uitgevoerd door een computer.
Dit lijkt misschien allemaal heel bijzonder maar is in de kern erg simpel (en dat zullen we je in deze cursus laten zien ;).
Programmeren is het opschrijven van instructies die samen een programma vormen.

Een programma kun je vervolgens uitvoeren en hergebruiken.
Soms heeft een programma ___input___ nodig, welke het nodig heeft om berekeningen te doen.
Soms leverd een programma ___output___, die door de gebruiker gebruikt kan worden.

Denk bijvoorbeeld aan een programma dat controleerd of je al bier mag kopen in de supermarkt.
Dit programma krijgt als input jouw geboortedatum en geeft als output ja of nee terug.

___Note:___ _In les 2 gaan we zo'n programma zelf maken ;)_.

Een programma wordt door een computer instructie voor instructie uitgevoerd.
Het begint dus bij stap 1 en gaat dan naar stap 2, stap 3 ... tot de laatste stap is berijkt.
Tussendoor kan het programma output genereren door dit te ___printen___ (hier beginnen we straks mee in les 1).

Een ___programma___ wordt ook wel een ___applicatie___, ___app___, ___toepassing___, ___script___ etc. genoemd.
Een programma bestaat uit ___code___, ___instructies___, ___regels___ etc.
Zo kun je vaak concepten vershillende namen geven.
Wij zullen proberen om in deze cursus dezelfde woorden te gebruiken zodat jij het makkelijk kan berijpen.
Maar het kan dus voorkomen dat je in deze cursus of op het internet andere woorden vind.
Het is belangrijk dat je het concept begrijpt, als je dan een andere cursus doet zal je het snel genoeg oppakken.

Een programma is dus een set van instructies die samen iets nutigs uitvoeren.
Die instructies kunnen door een computer worden gelezen.
Maar wat zijn nou deze instructies?
Op het allerlaagste niveau zijn dit nulletjes en eentjes (binair).

Zodat je het binaire programma later kunt uitvoeren.
En er zijn ___script talen___ die door zogeheten interpeters, een programma dat tekst omzet naar instructies die meteen uitgevoerd worden.

Een programmeertaal bestaat net als mensentaal uit taalregels/gramatica.
Dit noemen wij ___syntax___.
Als de syntax niet klopt kan het programma niet uitgevoerd worden en krijg je een foutmelding.
De foutmelding vertelt je dan wat je fout doet.

In deze cursus gaan wij je de basis concepten van syntax leren.
Elke taal heeft weer een eigen unieke syntax maar de concepten blijven hetzelfde.

## Hoe gaan we programmen

Je leest deze tekst in een webapplicatie (een programma in de vorm van een website) met de naam ___Jupyter Notebook___, wij noemen het gewoon ___Jupyter___.
Jupyter kan code blok voor blok uitvoeren.
Wij gebruiken de programmeertaal ___Python___, dus de code blokken bestaan uit Python code.
Zo kun je een paar python instructies opschrijven en deze daarna meteen testen.

Er zijn in Jupyter in principe twee soorten blokken.
Tekst blokken, waarin opgemaakte tekst staat, je kijkt er op dit moment naar ;).
Wij zullen dit ___Tekst blokken___ noemen.
Code blokken, waarin je code kunt uitvoeren.
Wij zullen dit ___Cellen___ of ___Code cellen___ noemen.

In een code cell kun je code schrijven en deze uitvoeren.
Bovenin deze pagina zie je een _"run"_ knop.
Als je in de cell hieronder selecteert door erop te klikken, kun je deze uitvoeren door op de _"run"_ knop te drukken.
Je kunt ook een sneltoets gebruiken.
Als je de cell geselecteert hebt kun je met _ctrl+Enter_ de cell uitvoeren.

Probeer het maar eens met de cell hieronder.
Klik erop en zie dat het kader rond de cell groen wordt.
Nu is de cell geselecteerd en kun je hem uitvoeren.


````python
1+1
````

Als je het goed doet zie je het antwoord van de rekensom (1+1) onder de cell verschijnen.

In deze cursus zal je telkens een stukje uitleg krijgen in een tekst blok met soms voorbeeld code in een code cell.
Daarna is er een opdracht met een code cell waarin je de opdracht kan uitwerken.

Om met de code cell te kunnen werken zal je een beetje moeten leren hoe je met Jupyter moet/kan omgaan.
Zoals je al gezien hebt zitten er knoppen bovenin de pagina.
Met deze knoppen kun je cellen toevoegen, verwijderen, uitvoeren, van positie wijzichen etc.

Als je een cell uitvoerd zie je links van de cell een nummer tussen twee blokhaken.
Dit nummer telt op als je een cell uitvoert.
Zo kun je zien welke cellen je in welke volgorde hebt uitgevoerd.

Soms heb je veel regels code in een cell staan, het is dan handig om de nummering aan te zetten.
Dat kan door de een na laatste knop te gebruiken.
